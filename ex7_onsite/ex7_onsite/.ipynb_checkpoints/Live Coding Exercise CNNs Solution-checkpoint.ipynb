```python
batch_size_train = 64
batch_size_test = 100

train_set = MNIST('data', download=True, transform=transforms.ToTensor())
test_set = MNIST('data', train=False, transform=transforms.ToTensor())

train_loader = torch.utils.data.DataLoader(train_set, batch_size=batch_size_train, num_workers=4, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_set, batch_size=batch_size_test, num_workers=4, shuffle=True)
```

```python
examples = enumerate(test_loader)
batch_idx, (example_data, example_targets) = next(examples)

import matplotlib.pyplot as plt

fig = plt.figure()
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.tight_layout()
    plt.imshow(example_data[i][0], cmap='gray', interpolation='none')
    plt.title("Ground Truth: {}".format(example_targets[i]))
    plt.xticks([])
    plt.yticks([])
```

## Defining the Network

First, we need to define our network architecture. It will be this sequence:
1. A 5x5 convolution with 10 output channels.
2. Max-Pooling with kernel size 2.
3. A 5x5 convolution with 20 output channels.
4. Max-Pooling again.
5. A fully connected layer with 50 output channels.
6. Dropout
7. A fully connected layer with 10 output channels. (10 for the number of digits)

Every layer except the last two is followed by a ReLU. 

The documentation of the Conv2d layer might help you: https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html And of course feel free to use the entire PyTorch documentation, e.g. for Dropout. 

(If you have not watched the VGG lecture for tomorrow yet: Convert the output of a convolutional layer to the input of a fully connected layer by flattening the array.)

```python
import torch.nn as nn
import torch.nn.functional as F

class MNISTClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(320, 50)
        self.fc2 = nn.Linear(50, 10)
        
    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2))
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))
        x = x.view(-1, 320)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, training=self.training)
        x = self.fc2(x)
        return x
    
loss_fct = nn.CrossEntropyLoss()
```

## Training

In theory we can just copy the training code from last time, and be done. However, we will practice some more advanced training setup today. (That is completely unnecessary for this task but good to know.)

### Epochs

If you read about training neural networks, they normally report their training epochs instead of training iterations. One epoch refers to training on the entire training set once.

Write a method train that trains the network for one epoch. (Tip: use last weeks training as a reference)

```python
model_path = './results/model.pth'
opt_path = './results/optimizer.pth'

def train(epoch):
    network.train() # tells the network we are training now, affects behavior of dropout etc 
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = network(data)
        loss = loss_fct(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
            train_losses.append(loss.item())
            train_counter.append(
                (batch_idx*batch_size_train) + ((epoch-1)*len(train_loader.dataset)))
            torch.save(network.state_dict(), model_path)
            torch.save(optimizer.state_dict(), opt_path)
```

### Saving and Logging

Training networks for bigger tasks can take days. If something fails in the mean time, you don't want to loose your entire training process. Change the train() function such that it will save the lastest model and optimizer states (!) after log_interval batches. (Doing it in every iteration is too slow, doing it after every epoch might be too long.) Google the corresponding torch functions.

Two blocks down you can also see that we will keep track of our training performance in the lists train_losses, train_counter, test_losses, test_counter. If you are done with the first task, fill these lists with the current loss and number of trained on elements.

```python
# this code is basically the same as our evaluate_dataset function last week, except it also calculates the accuracy
# accurarcy = n_correct_pred / n_pred
def test():
    network.eval() # tells the network that we are testing now, complements .train()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            output = network(data)
            test_loss += loss_fct(output, target).item()
            pred = output.data.max(1, keepdim=True)[1]
            correct += pred.eq(target.data.view_as(pred)).sum()
        test_loss /= len(test_loader.dataset) / batch_size_test
        test_losses.append(test_loss)
        print('\nTest set: Avg. loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
            test_loss, correct, len(test_loader.dataset),
            100. * correct / len(test_loader.dataset)))
```

```python
import os.path

n_epochs = 3
learning_rate = 0.01
momentum = 0.5
log_interval = 10

network = MNISTClassifier()
optimizer = torch.optim.SGD(network.parameters(), lr=learning_rate, momentum=momentum)

train_losses = []
train_counter = []
test_losses = []
test_counter = [i*len(train_loader.dataset) for i in range(n_epochs + 1)]


continue_training = True
if continue_training and os.path.isfile(model_path) and os.path.isfile(opt_path):
    network_state_dict = torch.load(model_path)
    network.load_state_dict(network_state_dict)
    
    optimizer_state_dict = torch.load(opt_path)
    optimizer.load_state_dict(optimizer_state_dict)


test()
for epoch in range(1, n_epochs + 1):
    train(epoch)
    test()
    
```

```python
fig = plt.figure()
plt.plot(train_counter, train_losses, color='blue')
plt.scatter(test_counter, test_losses, color='red')
plt.legend(['Train Loss', 'Test Loss'], loc='upper right')
plt.xlabel('number of training examples seen')
plt.ylabel('negative log likelihood loss')
```
